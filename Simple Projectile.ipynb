```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt

# Defining constants
g = 9.8  # acceleration due to gravity (m/s^2)

# Function to calculate trajectory
def projectile_motion(v0, theta, num_points=1000):
    """
    v0: Initial velocity in m/s
    theta: Angle of projection in degrees
    num_points: Number of points to generate for the trajectory curve
    """
    # Convert angle to radians
    theta_rad = np.radians(theta)
    
    # Time of flight
    t_flight = 2 * v0 * np.sin(theta_rad) / g
    
    # Time values from 0 to t_flight
    t = np.linspace(0, t_flight, num_points)
    
    # Horizontal and vertical positions as a function of time
    x = v0 * np.cos(theta_rad) * t
    y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2
    
    return x, y

# Function to plot the trajectory
def plot_trajectory(v0, theta):
    """
    v0: Initial velocity in m/s
    theta: Angle of projection in degrees
    """
    x, y = projectile_motion(v0, theta)
    
    # Plotting the trajectory
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label=f'v0 = {v0} m/s, theta = {theta}Â°')
    plt.title('Projectile Motion')
    plt.xlabel('Horizontal Distance (m)')
    plt.ylabel('Vertical Distance (m)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
v0 = 50  # initial velocity in m/s
theta = 45  # angle of projection in degrees

plot_trajectory(v0, theta)
```
